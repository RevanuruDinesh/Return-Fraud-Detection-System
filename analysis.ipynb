```python
import csv
from datetime import datetime

data = []

# Load CSV file
with open("../data/dataset.csv", mode='r') as file:
    reader = csv.reader(file)

    # Get headers
    headers = next(reader)
    print("Columns:", headers)

    # Read all rows
    for row in reader:
        data.append(row)

# Convert to structured dictionary format
structured_data = []

for row in data:
    row_dict = dict(zip(headers, row))
    structured_data.append(row_dict)

# Convert data types
for row in structured_data:
    try:
        row['price'] = float(row['price'])
        row['quantity'] = int(row['quantity'])
        row['discount'] = float(row['discount'])
    except:
        pass

# Convert date format
for row in structured_data:
    try:
        row['order_date'] = datetime.strptime(row['order_date'], "%d/%m/%Y")
        row['delivery_date'] = datetime.strptime(row['delivery_date'], "%d/%m/%Y")
    except:
        pass

# Create new feature (delivery gap)
for row in structured_data:
    try:
        row['delivery_gap'] = (row['delivery_date'] - row['order_date']).days
    except:
        row['delivery_gap'] = None

# Show first 5 records
for i in range(5):
    print(structured_data[i])
```

```python
import random

# Add new features
for row in structured_data:
    
    # Simulate return (30% chance)
    row['Return_Status'] = random.choice(['Yes', 'No', 'No'])
    
    # Simulate damage or wrong product
    row['Is_Damaged'] = random.choice(['Yes', 'No', 'No'])
    row['Is_Wrong_Product'] = random.choice(['Yes', 'No', 'No'])
    
    # Simulate customer history
    row['Total_Orders'] = random.randint(5, 20)
    row['Total_Returns'] = random.randint(0, row['Total_Orders'])
    
    # Calculate return ratio
    row['Return_Ratio'] = row['Total_Returns'] / row['Total_Orders']
    
    # FRAUD LOGIC (your idea)
    
    # If damaged or wrong → genuine
    if row['Is_Damaged'] == 'Yes' or row['Is_Wrong_Product'] == 'Yes':
        row['Fraud_Label'] = 0
    
    # High return ratio → fraud
    elif row['Return_Ratio'] > 0.5:
        row['Fraud_Label'] = 1
    
    # Fast return → possible fraud
    elif row['delivery_gap'] is not None and row['delivery_gap'] <= 2:
        row['Fraud_Label'] = 1
    
    else:
        row['Fraud_Label'] = 0

# Show results
for i in range(5):
    print(structured_data[i])
```

```python
# Prepare data for ML

X = []
y = []

for row in structured_data:
    
    # Convert Yes/No to numbers
    is_damaged = 1 if row['Is_Damaged'] == 'Yes' else 0
    is_wrong = 1 if row['Is_Wrong_Product'] == 'Yes' else 0
    
    return_status = 1 if row['Return_Status'] == 'Yes' else 0
    
    # Encode category (simple encoding)
    category_map = {
        'Electronics': 1,
        'Clothing': 2,
        'Home': 3,
        'Grocery': 4,
        'Beauty': 5
    }
    
    category = category_map.get(row['category'], 0)
    
    # Feature vector
    features = [
        row['price'],
        row['quantity'],
        row['discount'],
        row['delivery_gap'] if row['delivery_gap'] is not None else 0,
        row['Return_Ratio'],
        is_damaged,
        is_wrong,
        return_status,
        category
    ]
    
    X.append(features)
    y.append(row['Fraud_Label'])

# Check output
print("First 5 X:", X[:5])
print("First 5 y:", y[:5])
```

```python
# Simple train-test split (manual)

split_index = int(0.8 * len(X))

X_train = X[:split_index]
X_test = X[split_index:]

y_train = y[:split_index]
y_test = y[split_index:]

# -----------------------------------
# SIMPLE MODEL: RULE-BASED LEARNING
# -----------------------------------

def simple_model(features):
    price, quantity, discount, delivery_gap, return_ratio, is_damaged, is_wrong, return_status, category = features
    
    # Fraud logic (learned pattern)
    if is_damaged == 1 or is_wrong == 1:
        return 0
    
    if return_ratio > 0.5:
        return 1
    
    if delivery_gap <= 2:
        return 1
    
    return 0

# -----------------------------------
# PREDICTIONS
# -----------------------------------

y_pred = []

for features in X_test:
    pred = simple_model(features)
    y_pred.append(pred)

# -----------------------------------
# ACCURACY CALCULATION
# -----------------------------------

correct = 0

for i in range(len(y_test)):
    if y_test[i] == y_pred[i]:
        correct += 1

accuracy = correct / len(y_test)

print("Accuracy:", accuracy)

# -----------------------------------
# CLASSIFICATION REPORT (MANUAL)
# -----------------------------------

tp = fp = tn = fn = 0

for i in range(len(y_test)):
    if y_test[i] == 1 and y_pred[i] == 1:
        tp += 1
    elif y_test[i] == 0 and y_pred[i] == 0:
        tn += 1
    elif y_test[i] == 0 and y_pred[i] == 1:
        fp += 1
    elif y_test[i] == 1 and y_pred[i] == 0:
        fn += 1

# Precision, Recall, F1
precision = tp / (tp + fp) if (tp + fp) != 0 else 0
recall = tp / (tp + fn) if (tp + fn) != 0 else 0
f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

print("\nManual Classification Report:")
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
# SMART REFUND SYSTEM

def refund_decision(features):
    price, quantity, discount, delivery_gap, return_ratio, is_damaged, is_wrong, return_status, category = features
    
    # Rule 1: Genuine cases
    if is_damaged == 1 or is_wrong == 1:
        return "Full Refund ✅"
    
    # Risk scoring
    score = 0
    
    if return_ratio > 0.5:
        score += 2
    elif return_ratio > 0.3:
        score += 1
        
    if delivery_gap <= 2:
        score += 1
        
    if return_status == 1:
        score += 1
    
    # Decision
    if score <= 1:
        return "Full Refund ✅"
    elif score == 2:
        return "10% Deduction ⚠️"
    else:
        return "Manual Verification 🚨"

# TEST SYSTEM

print("SMART REFUND DECISIONS:\n")

for i in range(100):
    decision = refund_decision(X[i])
    print(f"Customer {i+1}: {decision}")
```

```python
import csv

# Save results
with open("../data/final_output.csv", mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Header
    writer.writerow([
        "order_id", "price", "return_ratio", "delivery_gap",
        "is_damaged", "is_wrong", "decision"
    ])
    
    # Data
    for i in range(len(X)):
        decision = refund_decision(X[i])
        
        order_id = f"ORD{i+1:05d}"   # Generates ORD00001, ORD00002...

        writer.writerow([
            order_id,
            X[i][0],  # price
            X[i][4],  # return_ratio
            X[i][3],  # delivery_gap
            X[i][5],  # is_damaged
            X[i][6],  # is_wrong
            decision
        ])

print("File saved successfully!")
```
